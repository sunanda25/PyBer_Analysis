```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
file_to_load= "Resources/got.csv"
got_df = pd.read_csv(file_to_load)
got_df
```

```python
# Get attacker and defender king data
king_attacker = got_df["attacker_king"].value_counts()
king_defender = got_df["defender_king"].value_counts()
king_attacker
```

```python
# Get total battle data
king_attacker_defender = king_attacker.add(king_defender, fill_value =0) 
king_attacker_defender
```

```python
# Configure plot and ticks

king_attacker_defender.plot(kind = "bar", figsize = (6, 5), color = "red")
# Set textual properties
plt.title("The Bloodthirst of Kings")
plt.xlabel("King")
plt.ylabel("Number of Battles Participated in")
plt.xticks(rotation=45)
# Show plot
plt.show()
# Resize plot to display labels
plt.tight_layout()
```
